**1.2 Perceptron**

A seperable boundray is solved using the perceptron algorithn on the given Dataset. 

```python
import numpy as np
from matplotlib import pyplot as plt

X = np.array([
    [1, 1, 1],
    [1, 2,-2],
    [1, -1,-1.5],
    [1, -2,-1],
    [1,-2, 1],
    [1, 1.5,-0.5],
   # [0, 3, 3]   # The extra data point
])
y = np.array([1, -1, -1, -1, 1, 1])
# The last label is the extra label which causes the perceptron to never converge
pts = 6

def predict(w, ip):
    
    s = np.dot(ip[1:], w[1:]) + w[0]
    if s >= 0:
        act = 1
    else:
        act = -1            
    
    return act

def Perceptron(ip, lab, n):

    w = np.zeros(ip.shape[1])
    w[0] = 0
    flag = 0
    k = 0
    while(flag != pts):
        error = []
        flag = 0
        for i, x in enumerate(ip):
            pred = predict(w, ip[i])
            error = lab[i] - pred
            w += n * (lab[i] - pred) * ip[i]
            if (error == 0):
                flag += 1
        print('Epoch:- ') 
        print(k)
        print(w)
        k += 1
        
        for j in range(0,pts):
    # Plot the negative samples
            if lab[j] == 1:
                plt.scatter(ip[j][1], ip[j][2], s=120, marker='+', linewidths=2)
    # Plot the positive samples
            else:
                plt.scatter(ip[j][1], ip[j][2], s=120, marker='_', linewidths=2)

        x = np.linspace(-4,4,100)

        l = (-w[1]*x-w[0])/w[2]

        plt.plot(x, l, 'r--', alpha = 0.9, label = 'y=mx+c')
                   
    return w, flag, k

wt = np.zeros(X.shape[1] + 1)

wt, con, epoch = Perceptron(X, y, 0.02)

if(con == pts):
    print('Converged on')

print(wt)

x = np.linspace(-4,4,100)
l = (-wt[1]*x-wt[0])/wt[2]
plt.plot(x, l, 'b--', alpha = 0.9, label = 'y=mx+c')
plt.title('Perceptron')
plt.xlabel('x1', color = '#1C2833')
plt.ylabel('x2', color = '#1C2833')
```

**2 Support Vector Machine**

The smae dataset used for perceptron is now solved using SVM. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

X = np.array([
    [1, 1],
    [2,-2],
    [-1,-1.5],
    [-2,-1],
    [-2, 1],
    [1.5,-0.5]
])

y = np.array([1,-1,-1,-1,1,1])

# fit the model, don't regularize for illustration purposes
svclassifier = SVC(kernel='linear')
svclassifier.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svclassifier.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(svclassifier.support_vectors_[:, 0], svclassifier.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```
